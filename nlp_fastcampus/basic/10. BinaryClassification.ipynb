```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.datasets import load_breast_cancer
cancer= load_breast_cancer()
```

```python
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df['TARGET'] = cancer.target

df.tail()
```

```python
df.describe()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
data  = torch.from_numpy(df.values).float()

data.shape
```

```python
x = data[:, :-1]
y = data[:, -1:]

x.shape, y.shape
```

```python
ratios = [.6, .2, .2]
```

```python
train_cnt = int(data.shape[0] * ratios[0])
valid_cnt = int(data.shape[0] * ratios[1])
test_cnt = data.shape[0] - train_cnt - valid_cnt
cnts = [train_cnt, valid_cnt, test_cnt]

cnts
```

```python
indices = torch.randperm(data.shape[0])

x = torch.index_select(x, dim=0, index=indices)
y = torch.index_select(y, dim=0, index=indices)

x = x.split(cnts, dim=0)
y = y.split(cnts, dim=0)

for x_i, y_i in zip(x, y):
    print(x_i.size(), y_i.size())
```

```python
scaler = StandardScaler()
scaler.fit(x[0].numpy())
x = [torch.from_numpy(scaler.transform(x[0].numpy())).float(),
     torch.from_numpy(scaler.transform(x[1].numpy())).float(),
     torch.from_numpy(scaler.transform(x[2].numpy())).float()]
```

```python
df = pd.DataFrame(x[0].numpy(), columns=cancer.feature_names)
df.tail()
```

```python
model = nn.Sequential(
    nn.Linear(x[0].shape[-1], 25),
    nn.LeakyReLU(),
    nn.Linear(25, 20),
    nn.LeakyReLU(),
    nn.Linear(20, 15),
    nn.LeakyReLU(),
    nn.Linear(15, 10),
    nn.LeakyReLU(),
    nn.Linear(10, 5),
    nn.LeakyReLU(),
    nn.Linear(5, y[0].shape[-1]),
    nn.Sigmoid()
)

model
```

```python
optimizer = optim.Adam(model.parameters())
```

```python
n_epochs = 10000
batch_size = 32
print_interval = 100
early_stop = 1000
```

```python
from copy import deepcopy

lowest_loss = np.inf
best_model = None

lowest_epochs = np.inf
```

```python
train_history, valid_history = [], []

for i in range(n_epochs):
    model.train()
    
    indices = torch.randperm(x[0].shape[0])
    x_ = torch.index_select(x[0], dim=0, index=indices)
    y_ = torch.index_select(y[0], dim=0, index=indices)
    
    x_ = x_.split(batch_size, dim=0)
    y_ = y_.split(batch_size, dim=0)
    
    train_loss, valid_loss = 0, 0
    y_hat = []
    
    for x_i, y_i in zip(x_, y_):
        y_hat_i = model(x_i)
        loss = F.binary_cross_entropy(y_hat_i, y_i)
        
        optimizer.zero_grad()
        loss.backward()
        
        optimizer.step()
        train_loss += float(loss)
        
    train_loss = train_loss / len(x_)
    
    with torch.no_grad():
        x_ = x[1].split(batch_size, dim=0)
        y_ = y[1].split(batch_size, dim=0)
        
        for x_i, y_i in zip(x_, y_):
            y_hat_i = model(x_i)
            loss = F.binary_cross_entropy(y_hat_i, y_i)
            
            valid_loss += float(loss)
            
            y_hat += [y_hat_i]
            
    valid_loss = valid_loss / len(x_)
    
    train_history += [train_loss]
    valid_history += [valid_loss]
    
    if (i + 1) % print_interval == 0:
        print('Epoch %d: train_loss=%.4e  valid_loss=%.4e  lowest_loss=%.4e'
              % (i + 1,
                 train_loss,
                 valid_loss,
                 lowest_loss))
        
    if valid_loss <= lowest_loss :
        lowest_loss = valid_loss
        lowest_epochs = i + 1
        
        best_model = deepcopy(model.state_dict())
        
    else :
        if (i + 1 - lowest_epochs) > early_stop and early_stop > 0 :
            print('No improvement %d epochs.' % early_stop)
            break
            
print("The Best validation loss from epoch %d: %.4e" % (lowest_epochs, lowest_loss))
model.load_state_dict(best_model)
```

```python
test_loss = 0
y_hat = []

with torch.no_grad():
    x_ = x[2].split(batch_size, dim=0)
    y_ = y[2].split(batch_size, dim=0)
    
    for x_i, y_i in zip(x_, y_):
        y_hat_i = model(x_i)
        loss = F.binary_cross_entropy(y_hat_i, y_i)
        
        test_loss += loss
        
        y_hat += [y_hat_i]

test_loss = test_loss / len(x_)
y_hat = torch.cat(y_hat, dim=0)

print('Test loss: %.4e' % test_loss)
```

```python
correct_cnt = (y[2] == (y_hat > .5)).sum()
total_cnt = float(y[2].size(0))

print('Test Accuracy : %.4f' % (correct_cnt / total_cnt))
```

```python
df = pd.DataFrame(torch.cat([y[2], y_hat], dim=1).detach().numpy(),
                  columns=['y', 'y_hat'])

sns.histplot(df, x='y_hat', hue='y', bins=50, stat='probability')
plt.show()
```
